```python
#unique values in each categorical colummn
def print_unique_col_values(df):
       for column in df:
            if df[column].dtypes=='object':
                print(f'{column}: {df[column].unique()}')
```

```python
print_unique_col_values(df1)
```

```python
#Some of the columns have no internet service or no phone service, that can be replaced with a simple No
df1.replace('No internet service','No',inplace=True)
df1.replace('No phone service','No',inplace=True)
```

```python
print_unique_col_values(df1)
```

```python
#to convert Yes or No columns to numeric
yes_no_columns = ['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup',
                  'DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']
for col in yes_no_columns:
    df1[col].replace({'Yes': 1,'No': 0},inplace=True)
```

```python
for col in df1:
    print(f'{col}: {df1[col].unique()}')
```

```python
#To convert gender to numeric
df1['gender'].replace({'Female':1,'Male':0},inplace=True)
```

```python
df1.gender.unique()
```

```python
#one hot encoding for columns with more than 2 categories
df2 = pd.get_dummies(data=df1, columns=['InternetService','Contract','PaymentMethod'])
df2.columns
```

```python
df2.sample(5)
```

```python
df2.dtypes
```

```python
#scles values with greater range to remove value bais that can occur
cols_to_scale = ['tenure','MonthlyCharges','TotalCharges']
#we use min-max scaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df2[cols_to_scale] = scaler.fit_transform(df2[cols_to_scale])
```

```python
for col in df2:
    print(f'{col}: {df2[col].unique()}')
```

```python
#spliting the dataset into train and test
X = df2.drop('Churn',axis='columns')
y = df2['Churn']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train[:10]
```

```python
len(X_train.columns)
```

**Build a model (ANN) in tensorflow/keras**

```python
import tensorflow as tf
from tensorflow import keras

model = keras.Sequential([
    keras.layers.Dense(26, input_shape=(26,), activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
    # Removed the optimizer from the layer list
])

# Optimizer is passed as an argument to model.compile
model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.01),
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=100)
```

```python
model.evaluate(X_test, y_test)
```

```python
yp = model.predict(X_test)
yp[:5]
```

```python
y_pred = []
for element in yp:
    if element > 0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
y_pred[:10]
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test,y_pred))
```

```python
import seaborn as sn
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_pred)

plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
y_test.shape
```

```python
# Calculate metrics: F1-score, recall, accuracy, binary crossentropy
report = classification_report(y_test, y_pred, output_dict=True)
f1_score = report['weighted avg']['f1-score']  # Extract F1-score from 'weighted avg'
recall = report['weighted avg']['recall']      # Extract recall from 'weighted avg'
accuracy = report['accuracy']                   # Accuracy is directly under the main dictionary

# Convert y_pred to a NumPy array
y_pred_array = np.array(y_pred)

# Calculate Binary Crossentropy with NumPy array
binary_crossentropy = tf.keras.losses.BinaryCrossentropy(from_logits=True)(y_test, y_pred_array).numpy().mean()

# Create a pandas DataFrame to store the metrics
data = {'Predicted Values': y_pred,
        'F1-Score': f1_score,
        'Recall': recall,
        'Accuracy': accuracy,
        'Binary Crossentropy': binary_crossentropy}
metrics_df = pd.DataFrame(data)
```

```python
# Save the DataFrame as a CSV file
metrics_df.to_csv('model_metrics.csv', index=False)

print("DataFrame saved as model_metrics.csv")
```
